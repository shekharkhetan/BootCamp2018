```python
#solution 1(part)
import numpy as np
from matplotlib import pyplot as plt

def MV_n():#if i want user to provide input i cannot take it as an argument#
    n= input("enter the number for which you need to create a n*n array:")   
    n = int(n)
    narray = np.random.normal(size=(n,n))
    xcv = np.mean(narray, axis =1)
    yv = np.var(xcv)
    return yv
#y = MV_n()  #printing a function by calling it
#print(y)
ar_mean_var = MV_n()
print("variance of the mean is:",ar_mean_var)
```

```python
#solution 1
from numba import jit 
@jit
def MV_n(n):
    narray = np.random.normal(size=(n,n))
    xcv = np.mean(narray, axis =1)
    yv = np.var(xcv)
    return yv
@jit
def arr_first():
    x = np.empty(10)
    n = np.array(range(100,1001,100))
    for i in range(10):
        x[i] = MV_n(n[i])
    plt.plot(n,x)
    

arr_first()
```

```python
#solution 2
def plot():
    x = np.linspace(-np.pi, np.pi, 201)
    plt.plot(x, np.sin(x))
    plt.xlabel('Angle [rad]')
    plt.ylabel('sin(x)')
    plt.axis('tight')
    plt.show();
    
    plt.plot(x,np.cos(x))
    plt.xlabel('Angle [rad]')
    plt.ylabel('cos(x)')
    plt.axis('tight')
    plt.show();
    
    plt.plot(x,np.arctan(x))
    plt.xlabel('Angle [rad]')
    plt.ylabel('arctan(x)')
    plt.axis('tight')
    plt.show()
plot()
```
