```python
#solution 1(part)
import numpy as np
from matplotlib import pyplot as plt

def MV_n():#if i want user to provide input i cannot take it as an argument#
    n= input("enter the number for which you need to create a n*n array:")   
    n = int(n)
    narray = np.random.normal(size=(n,n))
    xcv = np.mean(narray, axis =1)
    yv = np.var(xcv)
    return yv
#y = MV_n()  #printing a function by calling it
#print(y)
ar_mean_var = MV_n()
print("variance of the mean is:",ar_mean_var)
```

```python
#solution 1(full)
from numba import jit 
@jit
def MV_n(n):
    narray = np.random.normal(size=(n,n))
    xcv = np.mean(narray, axis =1)
    yv = np.var(xcv)
    return yv
@jit
def arr_first():
    x = np.empty(10)
    n = np.array(range(100,1001,100))
    for i in range(10):
        x[i] = MV_n(n[i])
    plt.plot(n,x)
    

arr_first()
#if we take range(100, 10000) the curve is smoother.
```

```python
#solution4
x = np.linspace(0,(2*np.pi),100)
fig, axes = plt.subplots(2,2)
plt.axis([0, 2*np.pi, -2, 2])
plt.tight_layout()
plt.suptitle("Mapping Sine functions", fontsize = 12)
axes[0,0].plot(x, np.sin(x), 'g-',)
axes[0,1].plot(x, np.sin(2*x), 'r--')
axes[1,0].plot(x, (2*np.sin(x)), 'b--')
axes[1,1].plot(x, (2*np.sin(2*x)), 'm:')
axes[0,0].set_title("Sin(x)", fontsize=10)
axes[0,1].set_title("Sin(2x)", fontsize=10)
axes[1,0].set_title("2*Sin(x)", fontsize=10)
axes[1,1].set_title("2*Sin(2x)", fontsize=10)
plt.tight_layout()
plt.show()
```

```python
x = np.linspace(-np.pi, np.pi, 100) 
y = x.copy() 
X, Y = np.meshgrid(x, y) 
Z = np.sin(X) * np.sin(Y) 

plt.subplot(131) 
plt.pcolormesh(X, Y, Z, cmap="viridis") 
plt.colorbar() 
plt.xlim(-np.pi, np.pi) 
plt.ylim(-np.pi, np.pi)
plt.tight_layout()
plt.show()

plt.subplot(132) 
plt.contour(X, Y, Z, 10, cmap="coolwarm") 
plt.colorbar()

plt.subplot(133) 
plt.contourf(X, Y, Z, [-1, -.8, -.5, 0, .5, .8, 1], cmap="magma") 
plt.colorbar()
plt.tight_layout()
plt.show()
```

```python
x = np.linspace(-2*np.pi, 2*np.pi, 100) 
y = x.copy() 
X, Y = np.meshgrid(x, y) 
Z = ((np.sin(X) * np.sin(Y))/(X*Y)) 

plt.subplot(121) 
plt.pcolormesh(X, Y, Z, cmap="viridis") 
plt.colorbar() 
plt.xlim(-2*np.pi, 2*np.pi) 
plt.ylim(-2*np.pi, 2*np.pi)
plt.tight_layout()
plt.show()

plt.subplot(122) 
plt.contour(X, Y, Z, 15, cmap="coolwarm") 
plt.colorbar()
```
