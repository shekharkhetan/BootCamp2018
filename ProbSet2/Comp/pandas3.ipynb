```python
#problem 1
import pandas as pd
from matplotlib import pyplot as plt
from pydataset import data
import numpy as np

iris = data('iris')
set(iris['Species'])
species=iris.groupby('Species')
species[['Sepal.Length', 'Sepal.Width', 'Petal.Length',
         'Petal.Width', 'Species']].mean().plot(kind='barh', title="Average  Sepal and Petal Properties")
plt.xlabel('Length in cm')
plt.ylabel('Petal Species')
print('1. setosa is the easiest to distinguish since it has very small average petal width  and length compared to other species')
iris


iris.plot(kind='hist', y='Sepal.Length')
iris.plot(kind='hist', y= 'Sepal.Width')
iris.plot(kind='hist', y='Petal.Length')
iris.plot(kind='hist', y='Petal.Width')

```

```python
poisons = data('poisons')
poisons
poisontype=poisons.groupby('poison')
treatment=poisons.groupby('treat')
poisontype[['time']].mean().plot(kind='barh', title= 'Average time for different poisons to act')
plt.xlabel('seconds')
print('Poison 3 is most deadly')
treatment[['time']].mean().plot(kind='barh', title= 'Average time for poison to act under treatment')
plt.xlabel('seconds')
print('Treatment D is most effective')

for i,j in poisontype:
    print(j.groupby('treat').mean())

```

```python
diamonds=data('diamonds')
colorgroup=diamonds.groupby('color')
cutgroup=diamonds.groupby('cut')
colorgroup['price'].mean().plot(kind='barh', title='Average price by color of diamonds')
plt.xlabel('price')
plt.tight_layout()
plt.show()
cutgroup['price'].mean().plot(kind='barh', title='Average price by cut of diamonds')
plt.xlabel('price')
plt.tight_layout()
plt.show()

```
