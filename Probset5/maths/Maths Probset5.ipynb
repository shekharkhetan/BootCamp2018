```python
import numpy as np
import scipy as scp
import matplotlib.pyplot as plt
```

```python
 # Exercise 8.1

x = np.linspace(0,50,1000)

y1 = (2*x+4)/3
y2 = (x-1)/6
y3 = 6-x
plt.plot(x,y1 ,'r', label='3y - 2x = 4')
plt.plot(x, y2,'g', label = 'x - 6y = 1')
plt.plot(x, y3, 'b', label = ' x + y = 6')
plt.plot(x, (5*x/4),'m--', label = "objective function 5x - 4y = 0")
plt.plot(x, (5*x-2)/4,'m--', label = "objective function 5x- 4y = 2")
y4 = np.maximum(y1,y2)
y5 = np.maximum(y3,y4)
plt.title('Maximization problem')
plt.fill_between(x,y5,y4,color = 'grey',alpha = 0.5, label='feasible region')
plt.xlim(0,8)
plt.ylim(0,8)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend(loc=4)
```

We know from the fundamental theorem of linear optimization that the optimal will lie on a vertex.
In the above question,we can see that there is only one vertex for the feasible set and the hyperplanes given in the problem. Therefore the solution will be intersection of the blue line ($x+y =6$) and red line ($2x-3y =-4$). This is given by the point $\mathbf{x = \frac{14}{5} ,y = \frac {16}{5}}$. ```python

```python
# Exercise 8.2(i)

x = np.linspace(0,50,1000)
y1 = (15-x)/3
y2 = (18-2*x)/3
y3 = x-4
plt.plot(x,y1 ,'r', label ='$3x_2 + x_1 = 15$')
plt.plot(x, y2,'g', label ='$3x_2+ 2x_1 = 18$')
plt.plot(x, y3, 'b', label = '$x_1 - x_2 = 4$')
plt.plot(x,10 -3*x,'m--', label =' objective fn.')
plt.plot(x,20 -3*x,'m--', label = 'objective function')

plt.title('Maximization problem')
y4 = np.minimum(y1,y2)
y5 = np.minimum(y3,y4)
plt.fill_between(x,y4,y5,color = 'magenta',alpha = 0.5)
plt.xlim(0,15)
plt.ylim(0,12)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend(loc=4)
print('The value of the objective function is 20')
print('The objective function is maximised at (6,2)')
```

We know from the fundamental theorem of linear optimization that the optimal will lie on a vertex. In the above question,we can see that there are two vertices for the feasible set and the hyperplanes given in the problem. In the optimization $x_1$ has three times weight than $x_2$. Hence solution will lie where $x_1$> $x_2$, which is intersection of the blue line ( $ x_1 -  x_2 =4 $ ) and green line ( $ 2x_1 + 3x_2 = 18$ ). This is given by the point  $ \mathbf{x_1=6, x_2=2}$ .
